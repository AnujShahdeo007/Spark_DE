```python
#3. 
#arr=[1,2,1,3,2]
arr=["a","b","a","c","b"]
seen=set()
result=[]
for x in arr:
    if x not in seen:
        seen.add(x)
        result.append(x)
print(result)
```

```python
from collections import Counter
s="aabbcddc"
freq=Counter(s)
print(freq)
result=None
for ch in s:
    if freq[ch]==1:
        result=ch
        break
print(result)
```
