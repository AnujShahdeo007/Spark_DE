```python
rdd=sc.parallelize([1,2,3,4,5,6],2)
def processs_partition(index,partition):
    for x in partition:
        yield (index,x)
result=rdd.mapPartitionsWithIndex(processs_partition)
print(result.collect())
```

```python
rdd1=sc.parallelize([
            (1,"Gaurav"),
            (2,"Ravi"),
            (3,"Neha")
        ])

rdd2=sc.parallelize([
            (1,"India"),
            (2,"USA"),
            (4,"UK")
        ])

# rdd1=(id,name)
#rdd2=(id,country)

# rdd partiton --> Shuffle -> regroup by key -> pair values 
```

Inner Join


```python
rdd1.join(rdd2).collect()
```

leftouterJoin() - left join 

    - Keep all keys from left Rdd
    - If match exists -> show value 
    - If no match -> right value becomes NONE

```python
rdd1.leftOuterJoin(rdd2).collect()
```

```python
rdd1.rightOuterJoin(rdd2).collect()
# Keep all keys from right RDD 
```

```python
rdd1.fullOuterJoin(rdd2).collect()
# Keep all keys from both rdd 
```

```python
# cogroup()
# Instead of pairing values, Spark group values from both RDD into list 
#Output stsructure : (key,([values from rdd1],[values from rdd2]))
cg=rdd1.cogroup(rdd2)
```

```python
print(cg.collect())
```

```python
for key,(v1,v2) in cg.collect():
    print(key,list(v1),list(v2))
```
