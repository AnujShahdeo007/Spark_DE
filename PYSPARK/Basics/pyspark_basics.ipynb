# PySpark Basics
This notebook demonstrates basic operations in PySpark, including initializing a SparkSession, creating a DataFrame, and displaying its contents.

## 1. Import PySpark Libraries
Import the necessary PySpark modules, such as SparkSession from pyspark.sql.

```python
from pyspark.sql import SparkSession
```

## 2. Initialize SparkSession
Create a SparkSession object, which is the entry point to using DataFrame and SQL functionality in PySpark.

```python
spark = SparkSession.builder.appName("PySparkBasics").getOrCreate()
```

## 3. Create a Simple DataFrame
Create a small DataFrame using a Python list of tuples and define the schema with column names.

```python
data = [(1, "Alice"), (2, "Bob"), (3, "Cathy")]
columns = ["id", "name"]
df = spark.createDataFrame(data, columns)
```

## 4. Show DataFrame Contents
Display the contents of the DataFrame using the show() method.

```python
df.show()
```

## 5. Stop SparkSession
Stop the SparkSession to release resources after the computation is done.

```python
spark.stop()
```
